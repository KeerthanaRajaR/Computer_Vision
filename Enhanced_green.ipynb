```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image

# Load the images
img1 = Image.open(r"C:\Users\Keerthana R\Pictures\Saved Pictures\New folder\image.jpg")
img2 = Image.open(r"C:\Users\Keerthana R\Pictures\Saved Pictures\New folder\images.jpg")


img1 = np.array(img1)
img2 = np.array(img2)

# Function to enhance only green color by boosting the green channel
def enhance_green(image, green_boost=1.5):
    enhanced_img = np.copy(image)
    
    # Only boost the green channel
    enhanced_img[..., 1] = np.clip(enhanced_img[..., 1] * green_boost, 0, 255)  # Green channel
    
    return enhanced_img

# Apply the green enhancement to both images
enhanced_img1 = enhance_green(img1)
enhanced_img2 = enhance_green(img2)

# Plot original and enhanced images side by side
fig, ax = plt.subplots(2, 2, figsize=(14, 14))

# Original images
ax[0, 0].imshow(img1)
ax[0, 0].set_title("Original Image 1")
ax[0, 0].axis('off')

ax[0, 1].imshow(img2)
ax[0, 1].set_title("Original Image 2")
ax[0, 1].axis('off')

# Enhanced images
ax[1, 0].imshow(enhanced_img1.astype(np.uint8))
ax[1, 0].set_title("Enhanced Green Image 1")
ax[1, 0].axis('off')

ax[1, 1].imshow(enhanced_img2.astype(np.uint8))
ax[1, 1].set_title("Enhanced Green Image 2")
ax[1, 1].axis('off')

plt.show()
```
